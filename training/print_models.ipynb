```python
from tensorflow.keras.utils import plot_model
from tensorflow.keras.models import load_model

# wczytanie modelu
model = load_model("../model/2025-09-28-16:40:50/bilstm_model.keras")

plot_model(
    model,
    to_file="bilstm_model.png",  # lub .pdf
    show_shapes=True,                 # pokaże wymiary tensorów
    show_layer_names=True,            # pokaże nazwy warstw
    dpi=96,
    rankdir="TB" # rozdzielczość pliku
)

```
